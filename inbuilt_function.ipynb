asynchronous

```python
async def async_generator():
    for i in range(3):
        yield i

async_iter = aiter(async_generator())
```

```python
async def async_generator():
    for i in range(3):
        yield i

async_iter = aiter(async_generator())
try:
    while True:
        value = await anext(async_iter)
        print(value)
except StopAsyncIteration:
    pass
```

```python
class Example:
    def __init__(self):
        self.attribute = "example"

obj = Example()
attributes = dir(obj)
print(attributes)
```

```python
global_variable = "I am global!"

def print_global():
    print(globals()['global_variable'])

print_global()
```

```python
def example_function():
    local_variable = "I am local!"
    print(locals()['local_variable'])

example_function()
```

```python
class Example:
    def __init__(self):
        self.attribute = "example"

obj = Example()
object_dict = vars(obj)
print(object_dict)
```
