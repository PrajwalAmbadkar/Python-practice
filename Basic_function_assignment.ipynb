

Q WA function to take list as parameter and print all numbers of given list which are divisible by 11


```python
def fn(ll):
  lis_11=list()
  lis_11=[o for o in ll if o%11==0]
  return lis_11

l=[11,22,75,7,4,8,1,7,88]
print(fn(l))
```

Q WA function to take string as parameter
Check if string is palindrome are not
If palindrome return 1
if not palindrome return 0

```python
def palindrome(s):
  if s==s[::-1]:
    return 1
  else:
    return 0

string='nitin'
print(palindrome(string))
```

```python
def reverse_string(input_str):
    reversed_str = ''.join(reversed(input_str))
    return reversed_str

# Example usage:
original_str = "Hello, World!"
reversed_str = reverse_string(original_str)

print(f"Original String: {original_str}")
print(f"Reversed String: {reversed_str}")
```

```python
def palindrome(s):
  start=0
  end=len(s)-1

  while start < end:
    if s[start]!=s[end]:
      return False
    start+=1
    end-=1
  return True

a=palindrome('ramar')
if a:
  print('palindrome')
else:
  print('not palindrome')
```

Q WA function to check if number is prime or not. Return True if number is prime and false if number not prime.

```python
def prime(num):
  if num==1:
    return False
  elif num>1:
    for i in range(2,num):
      if num%i==0:
        return False
    else:
        return True

number=13
print(prime(number))
```

Q WA function to take a dictionary as parameter. Print value of key "name" if key "rollno" is having value 100

```python
def keyroll(d):
  for i,j in d.items():
    if j==100:
      print(i)

keyroll({'eva':100,'aadi':50})
```

Q WA function to list of strings as parameter and print all strings which have character 's' 2 times

```python
def fn1(l):
  for i in l:
    if i.count('s')==2:
      print(i)

ll=['eva','wall-e','ss','ss rajmauli']
fn1(ll)
```

Q WA function to take list of dictionaries. Print all dictionaries which contain key "id".


```python
def funct(l):
  for dicts in l:
    for key,value in dicts.items():
      if key=='id':
        print(dicts)


list_of_dicts = [
    {"name": "John", "id": 1, "age": 25},
    {"name": "Jane", "age": 30},
    {"id": 2, "city": "New York"},
    {"username": "user123", "email": "user@example.com"}]
funct(list_of_dicts)
```
