Q Sort all the numbers in a list in descending order using lambda function


```python
l=[4,6,8,1,3,2]
rev=sorted(l,key=(lambda x:x),reverse=True)
print(rev)
```

Q Sort all the characters of given string using lambda function


```python
ll=['c','h','s','i']
rev=sorted(ll,key=(lambda x:x),reverse=False)
rev
```

Q Given list of strings, sort all the strings by last character of that string. . Use lambda function and normal function both.

```python
string = ["apple", "banana", "cherry", "date", "fig"]
sort=sorted(string,key=(lambda x:x[-1]),reverse=False)
sort
```

Q Given list of employees. This list may contain repetitions. Find all unique employee names and print them as per order of second character in that name. Use lambda function and normal function both.

```python
employees = ["Alice", "Bob", "Charlie", "David", "Bob", "Alice", "Eve", "Charlie"]
set_emp=set(employees)
set_emp=sorted(set_emp,key=(lambda x:x[2]),reverse=False)
set_emp
```

```python
def sort_fn(emp):
  return emp[2]

employees = ["Alice", "Bob", "Charlie", "David", "Bob", "Alice", "Eve", "Charlie"]
set_emp=list(set(employees))
set_emp.sort(key=sort_fn)
set_emp
```

Q. Take Employee id and Employee name of 5 employees from user. Store it in dictionary. Print all employees in increasing order of employee Id . Also print all employees in alphabetical order by name.


```python
dic={3:'Charlie', 9:'Bob', 1:'Eve', 6:'Alice', 5:'David'}
ord_dic=dict(sorted(dic.items()))
ord_dic
```

```python
dic={3:'Charlie', 9:'Bob', 1:'Eve', 6:'Alice', 5:'David'}
ord_dic=dict(sorted(dic.items(),key=(lambda item:item[1])))
ord_dic
```

Q. Take comma separated numbers as input from the user. Split it in list of strings. Now convert every string in this list to float using map function

```python
l='4,6,8,1,3,2'
s_l=l.split(',')
sl=list(map(lambda x:float(x),s_l))
sl
```

```python
l='4,6,8,1,3,2'
s_l=l.split(',')
sl=list(map(float,s_l))
sl
```

Given list of floating point numbers. Convert every number into string and then join all the numbers such that they are separated by pipe(|)

```python
sl_str=list(map(str,sl))
str_sl="|".join(sl_str)
str_sl
```

Q Given list of integers. Use map function to calculate square of all and store in list. Then print sum of all square values.

```python
l=[4,6,8,1,3,2]
sq_l=list(map(lambda x:x**2,l))
print(sum(sq_l))
```
