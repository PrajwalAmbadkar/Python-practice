5. Take a sentence as input from user. Every word is seperated by space.
a. Create a word_count dictionary which will have unique words and their count.


```python
string='A regression is A statistical technique that relates a dependent variable to one or more'
lstring=string.lower().split()
wc={}
for word in lstring:
  if word in wc:
    wc[word]+=1
  else:
    wc[word]=1

for i,j in wc.items():
  if j==1:
    print(i,":",j,end=" , ")
```

```python

string='A regression is A statistical technique that relates a dependent variable to one or more'
lstring=string.lower().split()
wc={}
distint_l=list(set(lstring))

for i in distint_l:
  wc.update({i:lstring.count(i)})
print(wc)
```

b. create suffix_count dictionary which will contain count of words ending with 's', 'es', 'ed', 'y', 'en' , etc.


```python
def fun(l,sub):
  sam=0
  for j in l:
    if j.endswith(sub):
        sam+=1
  return sam

string='A regression is A statistical technique that relates a dependent variable to one or more'
l=string.lower().split()

l1=[ 's', 'es', 'ed', 'y', 'en' ]
dic={}
for i in l1:
  dic[i]=fun(l,i)
print(dic)
```

c. create dictionary word_length_count which will store length of word and count.
Ex.
input_sent= 'CDAC is in Pune'
There are 2 words of length 4 and 2 words of length 2 so,
word_length_count = {2:2, 4:2}

```python
word_length_count={}
input_sent= 'CDAC is in Pune'
l=input_sent.lower().split()
for i in l:
  if len(i) in word_length_count:
    word_length_count[len(i)]+=1
  else:
    word_length_count[len(i)]=1


print(word_length_count)
```
